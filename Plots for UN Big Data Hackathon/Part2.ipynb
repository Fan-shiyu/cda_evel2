### Rainfall

```r
path <- '/home/shiyu/Desktop/UN contest/data/d/'
fig_path <- '/home/shiyu/Desktop/UN contest/figs/'
```

```r
d <- read.csv(paste0(path, 'jsl2010-2021.csv'))
dim(d)
head(d, 2)
```

```r
pdf(paste0(fig_path, 'Chinese Rainfall (from 2010 to 2021).PDF'), 
    width = 16, height = 10)
cols <- c('#fbb4ae', '#b3cde3', '#ccebc5')
at <- rep(1:3, 12) + rep(0:11,each=3) * 4
boxplot(d$pre ~ d$ym, pch=20, cex=0.5, col=cols, xlim=c(-3.4,50), 
        axes=FALSE, xlab='', ylab='', at=at)
abline(v=-1, col=grey(0.2,0.9))
lines(c(-1,52), rep(-220,2), col=grey(0.2,0.9))

mtext('Chinese Rainfall (from 2010 to 2021)',side=3, line=1, cex=1.2)
mtext('-- in July, August, September',side=3, line=-0.4, cex=1)
mtext('Year', side=1, line=1.5)
mtext('Monthly Rainfall', side=2, line=-2)
mtext(seq(2010,2021,by=2), side=1, line=0.3, at=seq(2,by=8,l=6))

text(-2.2, seq(0,5500, by=500), seq(0,5500, by=500))
legend(43, 6000, substr(month.name, 1, 3)[7:9], 
       box.col=grey(0.2,0.8), box.lwd=0.6, pch=15, cex=1.2,
       lty=0, lwd=3, col=cols, x.intersp=0.1, y.intersp=0.8)
dev.off()
```
